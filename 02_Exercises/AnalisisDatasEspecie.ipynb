### *Datos Especie y peso animal*

   - record_id: Identificador único de la observación
   - month: Mes de observación
   - day: Día de la observación
   - year: Año de la observación
   - plot_id: ID de un sitio en particular
   - species_id: Código de dos letras
   - sex: Sexo del animal (“M”, “F”)
   - hindfoot_length: Tamaño de pata en mm
   - weight: Peso del animal en gramos

```python
# Agrupación por la columna 'sex' y calculo de su media en cada columna

gb_sex = dfE.groupby('sex')
gb_sex.mean()
```

```python
# Agrupación por dos caracteres 'plot_id' y 'sex' (Media según el sexo en cada lugar)

gb_2 = dfE.groupby(['plot_id','sex'])
gb_2.mean()
```

```python
# Calculamos la estadística del peso para cada sitio. Solo representamos la columna 'weight'

gb_site = dfE.groupby('plot_id')
gb_site['weight'].describe()
```

```python
# Contamos el número de muestras para cada especie

species_count= dfE.groupby('species_id')['record_id'].count()
print(species_count)
```

```python
# Si solo queremos saber los registros de la especie 'DO'

species_do= dfE.groupby('species_id')['record_id'].count()['DO']
print(species_do)
```

```python
# Creamos una gráfica de barras
%matplotlib inline
species_count.plot.bar()
```
